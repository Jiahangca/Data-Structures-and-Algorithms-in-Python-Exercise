```python
# R-13.7
def compute_kmp_fail(P):
  m = len(P)
  fail = [0] * m
  j, k = 1, 0
  while j < m:
    if P[j] == P[k]:
      fail[j] = k + 1
      j += 1
      k += 1
    elif k > 0:
      k = fail[k-1]
      # k = fail[0]
    else:
      j += 1
  return fail

if __name__ == '__main__':
  P = 'cgtacgttcgtac'

  fail = compute_kmp_fail(P)
  print(fail)

  for k in range(len(P)):
    print("k: {0}, P(k): {1}, f(k): {2}".format(k, P[k], fail[k]))
  
# 1. in Python dictionary, one key is mapped to one value, but in this pattern string there are duplicated keys(characters), so we use format function rather than a dictionary in R-13.6
```

```python
# R-13.8
'''
the main idea is multiplying large matrix first
10*5, 5*2, 2*20, 20*12, 12*4, 4*60
10 and 60 cannot be consumed, for the others, 20 > 12 > 5 > 4 > 2
so 2*20*12 + 2*12*4 + 10*5*2 + 2*4*60 + 10*2*60 = 2356



'''
def matrix_chain(d):
  n = len(d) - 1
  N = [[0] * n for i in range(n)]
  for b in range(1, n):
    for i in range(n-b):
      j = i + b
      N[i][j] = min(N[i][k] + N[k+1][j] + d[i] * d[k+1] * d[j+1] for k in range(i, j))
  return N

if __name__ == '__main__':
  d = [10,5,2,20,12,4,60]
  N = matrix_chain(d)
  print(N[0][-1])
```

```python
# R-13.9
'''
another common subsequence with length six is CTAATA
'''
def LCS(X, Y):
  n, m = len(X), len(Y)
  L = [[0] * (m + 1) for k in range(n + 1)]
  for j in range(n):
    for k in range(m):
      if X[j] == Y[k]:
        L[j+1][k+1] = L[j][k] + 1
      else:
        L[j+1][k+1] = max(L[j][k+1], L[j+1][k])
  return L

def LCS_solution(X, Y, L):
  solution = []
  j, k = len(X), len(Y)
  while L[j][k] > 0:
    if X[j-1] == Y[k-1]:
      solution.append(X[j-1])
      j -= 1
      k -= 1

    # elif L[j-1][k] >= L[j][k-1]:
    elif L[j-1][k] > L[j][k-1]:
      j -= 1

    else:
      k -= 1
  return ''.join(reversed(solution))

if __name__ == '__main__':
  X = 'GTTCCTAATA'
  Y = 'CGATAATTGAGA'
  # Y = 'GTTCCTAATA'
  # X = 'CGATAATTGAGA'
  L = LCS(X, Y)
  print(L)
  print(LCS_solution(X, Y, L))

# 1. in the original code, when L[j-1][k] = L[j][k-1], which means neither the ends are included in the common sequence, j -= 1. However, we can also make it k -= 1, which leads to the other solution
```

```python
def find_brute(T, P):
  n, m = len(T), len(P)
  for i in range(n-m+1):
    k = 0
    while k < m and T[i+k] == P[k]:
      k += 1
    if k == m:
      return i
  return -1
```

```python
def find_boyer_moore(T, P):
  n, m = len(T), len(P)
  if m == 0:
    return 0
  last = {}
  for k in range(m):
    last[P[k]] == k
  i = m - 1
  k = m - 1
  while i < n:
    if T[i] == P[k]:
      if k == 0:
        return i
      else:
        i -= 1
        k -= 1
    else:
      j = last.get(T[i], -1)
      i += m - min(k, j+1)
      k = m - 1
  return -1

if __name__ == '__main__':
  P = 'JIAHANG'
  last = {}
  for k in range(len(P)):
    last[P[k]] = k
  print(last)
```

```python
def find_kmp(T, P):
  n, m = len(T), len(P)
  if m == 0:
    return 0
  fail = compute_kmp_fail(P)
  # j for text, k for pattern
  j, k = 0, 0
  while j < n:
    if T[j] == P[k]:
      if k == m - 1:
        return j-m+1
      j += 1
      k += 1
    elif k > 0:
      k = fail[k-1]
    else:
      j += 1
  return -1
```
