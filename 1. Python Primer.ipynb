```
# R-1.1

def is_multiple(n,m):
  if n % m == 0:
    return True
  else:
    return False

is_multiple(9,9)
```

```
# R-1.2

def is_even_1(k):
  if k ^ 1 == k + 1:
    return 'Even'
  else:
    return 'odd'

def is_even_2(k):
  if k & 1 == 0:
    return 'Even'
  else:
    return 'odd'

def is_even_3(k):
  if k | 1 == k + 1:
    return 'Even'
  else:
    return 'odd'

print(is_even_1(5))
print(is_even_1(6))
print(is_even_2(5))
print(is_even_2(6))
print(is_even_3(5))
print(is_even_3(6))

# odd or even depends on the rightest bit. in bit operator, 
# 1 is equal to 00001, all the higher bits can be considered as 0
```

```
# R-1.4

def sum_squares(n):
  # i = 0
  sum = 0
  for i in range(n):
    sum += i * i
  return sum

sum_squares(4)

# += and = a + b have subtle difference only in list class.
```

```
# R-1.12

import random
def choice(data):
  k = random.randrange(0,len(data),1)
  return data[k]

d = [1,2,3,5,7,9,11,13,17,19]
print(choice(d))
```

```
# C-1.28

def norm(v, p=2):
  i = 0
  sum = 0
  for i in range(len(v)):
    sum += v[i]**p
  return sum**(1/p)

v = [3,4]
norm(v,3)
```
