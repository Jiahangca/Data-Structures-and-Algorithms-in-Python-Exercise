```
# P-3.58

def unique1(S):
  """Return True if there are no duplicate elements in sequence S."""
  for j in range(len(S)):
    for k in range(j+1, len(S)):
      if S[j] == S[k]:
        return False              # found duplicate pair
  return True                     # if we reach this, elements were unique

def unique2(S):
  """Return True if there are no duplicate elements in sequence S."""
  temp = sorted(S)                # create a sorted copy of S
  for j in range(1, len(temp)):
    if S[j-1] == S[j]:
      return False                # found duplicate pair
  return True                     # if we reach this, elements were unique

from time import time
import numpy as np

S = np.random.permutation(int(1e3)) 
np.append(S, [5])

start_time_1 = time()
unique1(S)
end_time_1 = time()
elapsed_1 = end_time_1 - start_time_1
print(start_time_1, end_time_1, elapsed_1)

start_time_2 = time()
unique2(S)
end_time_2 = time()
elapsed_2 = end_time_2 - start_time_2
print(start_time_2, end_time_2, elapsed_2)

# 1. there are 2 ways to produce random numbers. the python build-in random module and numpy random module. random can only give a number but np.random can also give a numpy.ndarray (not list)
#   for numpy.random, read Python data science handbook
#   top functions in random module:
#     (1) random.seed(hashable): initilize the pseudo-random number generator based on the hash value of the parameter
#     (2) random.random(): no parameter, return a float belong to [0.0, 1.0)
#     (3) random.randint(low, high): return an integer belong to [low, high]
#     (4) random.uniform(a, b): return a float from [a,b] (if a > b, from [b, a])
#     (5) random.randrange(start, stop, step): return an integer from range(start, stop, step)
#     (6) random.choice(seq): retuen an element;  random.shuffle(seq): reorder the elements
# 2. time.time() return the time in seconds since the epoch as a floating point number. 
# 3. np.random.permutation() only accept an integer or array, 1e3 is a float, but can convert it to int.
```
