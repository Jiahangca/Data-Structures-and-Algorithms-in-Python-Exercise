```
# R-4.6
'''
Algorithm Harmonic(n):
Input: integer n
Output: the nth Harmonic number
if n == 1:
  return 1
else:
  return 1/n + Harmonic(n-1)
'''
def Harmonic(n):

  if n == 1:
    return 1
  else:
    return 1/n + Harmonic(n-1)


if __name__ == '__main__':
  print(Harmonic(3))
# 1. transformation between recursion and nonrecursion:
  # (1) control parameter: every paraeter change in the recursion procedure should be a formal parameter of the recursive function; is a loop test in nonrecursion
  # (2) return to base case: always have a base case; always focus on how does it move towards to the base case; always focus on the output especially for special situation(only have base case)
```

```
# R-4.7
'''
Algorithm StringToInteger(s, i=len(s)-1, integer=0):
Input: string sequence s, 
Output: the integer form of s
if i < 0 :
  return integer
else:
  integer + =s[len()-1-i] * 10 ** i
  StringToInteger(s, i-1, integer)
'''

def stringtointeger(s, i):
  k = len(s) - 1
  if i > 0:
    return int(s[k-i]) * 10 ** i + stringtointeger(s, i-1)
  elif i == 0:
    return int(s[k])


if __name__ == '__main__':
  s = '135789'
  i = len(s)-1
  print(stringtointeger(s=s, i=i))

# 1. string is also a sequence type (like tuple and list) and support s[i]
# 2. [0-9]e[0-9], the left and right of e should be a integer or float, k=1, 3ek is illegal; use 10**k
# 3. a function will return none when you use print() instead of return
# 4. in recursive function, it is okay to use a parameter to track elements, such as max in R-4.1. but the sum parameter (such as sum += in nonrecursive function) isn't needed 
```

```
# C-4.22
def power(x, n):
  """Compute the value x**n for integer n."""
  if n == 0:
    return 1
  else:
    partial = power(x, n // 2)          # rely on truncated division
    result = partial * partial
    if n % 2 == 1:                      # if n odd, include extra factor of x
      result *= x                       
    return result

def is_even_1(k):
  if k ^ 1 == k + 1:
    return 'Even'
  else:
    return 'odd'

def power_nonrecursive(x, n):
  counter = 0
  partial = 1
  _n = n

  while _n:
    _n = _n >> 1
    counter += 1

  while counter+1:
    # in clude the outest one x ^ n
    partial = partial * partial
    if is_even_1(n>>counter) == 'odd':
      # ture when n is odd, start from n = 0
      partial = partial * x
    counter -= 1

  return partial


if __name__ is '__main__':
  s = [(2, 13), (2, 18)]
  for x, n in s:
    # just like looping through a dictionary
    print(power_nonrecursive(x, n))

# 1. don't use formal parameter directly into calculation. use a inner identifier to ensure formal parameter is never changed!
# 2. 3 method to check odd and even: the & is the most straight forward
    # (1) 0 ^ 1 = 0 + 1;    1 ^ 1 = 0   ^ : result is 1 when left and right not same
    # (2) 0 & 1 = 0;       1 & 1 = 1
    # (3) 0 | 1 = 0 + 1;    1 | 1 = 1   | : result is 1 when left or right is 1
```
