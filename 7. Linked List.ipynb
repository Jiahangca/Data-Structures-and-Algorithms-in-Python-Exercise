```
# R-7.22,23


class FavoritesList:
  """List of elements ordered from most frequently accessed to least."""

  #------------------------------ nested _Item class ------------------------------
  class _Item:
    __slots__ = '_value', '_count'             # streamline memory usage
    def __init__(self, e):
      self._value = e                          # the user's element
      self._count = 0                          # access count initially zero

  #------------------------------- nonpublic utilities -------------------------------
  def _find_position(self, e):
    """Search for element e and return its Position (or None if not found)."""
    walk = self._data.first()
    while walk is not None and walk.element()._value != e:
      walk = self._data.after(walk)
    return walk

  def _move_up(self, p):
    """Move item at Position p earlier in the list based on access count."""
    if p != self._data.first():                      # consider moving...
      cnt = p.element()._count
      walk = self._data.before(p)
      if cnt > walk.element()._count:                # must shift forward
        while (walk != self._data.first() and
               cnt > self._data.before(walk).element()._count):
          walk = self._data.before(walk)
        self._data.add_before(walk, self._data.delete(p))   # delete/reinsert
  
  #------------------------------- public methods -------------------------------
  def __init__(self):
    """Create an empty list of favorites."""
    self._data = PositionalList()                 # will be list of _Item instances
    # print('self: ', type(self))
    # print('self._data: ', self._data)
    # print('self._data.first()', type(self._data.first()))

  def __len__(self):
    """Return number of entries on favorites list."""
    return len(self._data)

  def is_empty(self):
    """Return True if list is empty."""
    return len(self._data) == 0

  def access(self, e):
    """Access element e, thereby increasing its access count."""
    p = self._find_position(e)                    # try to locate existing element
    if p is None:
      p = self._data.add_last(self._Item(e))      # if new, place at end
    p.element()._count += 1                       # always increment count
    self._move_up(p)                              # consider moving forward

  def remove(self, e):
    """Remove element e from the list of favorites."""
    p = self._find_position(e)                    # try to locate existing element
    if p is not None:
      self._data.delete(p)                        # delete, if found 

  def top(self, k):
    """Generate sequence of top k elements in terms of access count."""
    if not 1 <= k <= len(self):
      raise ValueError('Illegal value for k')
    walk = self._data.first()
    for j in range(k):
      item = walk.element()                       # element of list is _Item
      yield item._value                           # report user's element
      walk = self._data.after(walk)

  def __repr__(self):
    """Create string representation of the favorites list."""
    return ', '.join('({0}:{1})'.format(i._value, i._count) for i in self._data)


  def clear(self):
    # remove() is a method in the FavoriteList class
    for i in range(len(self._data)):
      self.remove(self._data.first().element()._value)
    if len(self._data) == 0:
      return 'The list is empty now!'

  def reset_counts(self):
    p = self._data.first()
    while p != None:
      p.element()._count = 0
      p = self._data.after(p)




if __name__ == '__main__':
  fav = FavoritesList()
  for c in 'hello. this is a test of mtf':        # well, not the mtf part...
    fav.access(c)
    # k = min(5, len(fav))
    # print('Top {0} {1} {2}'.format(k, [x for x in fav.top(k)], fav))

  print(len(fav))
  print(fav)
  fav.reset_counts()
  print(fav)
  # print(fav.clear())
  # print(len(fav))


# 1. the access() method in the FavoritesList class only add non-repeating elements. len('hello. this is a test of mtf') = 28, but len(set('hello. this is a test of mtf')) = 12
# 2. before working on a class, the first is determining the object type
# 3. remove() is a method in the FavoriteList class, its formal parameter e is the _value parameter of _Item class
# 4. self._data is a PositionalList object, but it is an inner variable and cannot be accessed from outside. FavoritesList itself is an ordinary class
```

```
# C-7.43
class Empty(Exception):
  pass

class LinkedDeque(_DoublyLinkedBase):         # note the use of inheritance
  """Double-ended queue implementation based on a doubly linked list."""

  def first(self):
    """Return (but do not remove) the element at the front of the deque.

    Raise Empty exception if the deque is empty.
    """
    if self.is_empty():
      raise Empty("Deque is empty")
    return self._header._next._element        # real item just after header

  def last(self):
    """Return (but do not remove) the element at the back of the deque.

    Raise Empty exception if the deque is empty.
    """
    if self.is_empty():
      raise Empty("Deque is empty")
    return self._trailer._prev._element       # real item just before trailer

  def insert_first(self, e):
    """Add an element to the front of the deque."""
    self._insert_between(e, self._header, self._header._next)   # after header

  def insert_last(self, e):
    """Add an element to the back of the deque."""
    self._insert_between(e, self._trailer._prev, self._trailer) # before trailer

  def delete_first(self):
    """Remove and return the element from the front of the deque.

    Raise Empty exception if the deque is empty.
    """
    if self.is_empty():
      raise Empty("Deque is empty")
    return self._delete_node(self._header._next)   # use inherited method

  def delete_last(self):
    """Remove and return the element from the back of the deque.

    Raise Empty exception if the deque is empty.
    """
    if self.is_empty():
      raise Empty("Deque is empty")
    return self._delete_node(self._trailer._prev)  # use inherited method

def card_shuffle(L):
# L has 2n elements, assume it is a LinkedDeque inherited from _DoublyLinkedBase class
  length = len(L)
  l1 = LinkedDeque()
  l2 = LinkedDeque()
  l_final = LinkedDeque()
  for i in range(length//2):
    l1.insert_last(L.delete_first())
  for i in range(length//2):
    l2.insert_last(L.delete_first())

  for i in range(length//2):
    l_final.insert_last(l1.delete_first())

    l_final.insert_last(l2.delete_first())

  return l_final

if __name__ == '__main__':
  L = LinkedDeque()
  for i in range(1,7):
    L.insert_last(i)
  L_shuffle = card_shuffle(L)
  for i in range(len(L_shuffle)):
    print(L_shuffle.delete_first())

# 1. if access is more frequent, use array_based structure, if insertion and deletion are more frequent, use linked structure
# 2. 4/2 produce float, 4//2 produce int
# 3. in the third for loop, one execution of body contains two operations, so the interable is still range(length//2)
```

```
# P-7.46
%xmode verbose
class ArrayPositionalList:

  # component class _ArrayBase
  class _ArrayBase:
    __slots__ = '_value', '_index'
    def __init__(self, e):
      self._value = e
      self._index = None



  # composite class
  def __init__(self):
    # simplify the code
    self._data = [None] * 20
    self._n = 0
  


# -----------utility method-------------

  def _insert(self, i, e):
    new = self._ArrayBase(e)
    if self._n == 0 :
      self._data[0] = new
      self._n += 1
      self._data[0]._index = 0
    else:
      print('self._n: {0} i: {1}'.format(self._n, i))
      print('range(self._n, i, -1):', list(range(self._n, i, -1)))
      for k in range(self._n, i, -1):
        print('k:', k)
        self._data[k] = self._data[k-1]
        self._data[k]._index += 1
      self._data[i] = new
      self._data[i]._index = i
      self._n += 1

  def _delete(self, i):
    e = self._data[i]
    for k in range(i, self._n):
      self._data[k] = self._data[k+1]
      self._data[k]._index -= 1
    self._n -= 1
    return e

# ------------accessors---------------

  def first(self):
    if self._data[0] == None:
      raise Exception('the list is empty!')
    else:
      # print(type(self._data[0]))
      return self._data[0]._value

  def last(self):
    if self._n <= 0:
      raise Exception('the list is empty!')
    else:
      return self._data[self._n-1]._value

  def before(self, i):
    return self._data[i-1]._value


  def after(self, i):
    return self._data[i+1]._value

  def is_empty(self):
    return self._n == 0

  def __len__(self):
    return self._n

  def __iter__(self):
    i = 0
    while self._data[i] is not None:
      yield self._data[i]._value
      i += 1


# -----------mutators---------------

  def add_first(self, e):
    return self._insert(0, e)

  def add_last(self, e):
    # print(self._n)
    return self._insert(self._n, e)

  def add_before(self, i, e):
    return self._insert(i, e)

  def add_after(self, i ,e):
    return self._insert(i+1, e)
  
  def replace(self, i ,e):
    self._data[i]._value = e

  def delete(self, i):
    self._delete(i)


if __name__ == '__main__':
  L = ArrayPositionalList()

  L.is_empty()
  len(L)

  L.add_first(5)      #[5]
  print(list(L))
  L.add_last(3)       #[5, 3]
  print(list(L))

  L.add_before(1,4)     #[5, 4, 3]
  print(list(L))
 
  L.add_after(1,7)      #[5, 4, 7, 3]
  print(list(L))

  print(list(L))


  print(L.first())
  print(L.last())
  print(L.before(1))
  print(L.after(1))

  print(L.is_empty())
  print(len(L))

  P = iter(L)
  print(next(P))
  print(next(P))
  print(next(P))
  print(next(P))
# 1. positional list ADT is developed because we want a more general data type than stack, queue, and deque, which we can operate the items between the first and last. because link-based structure is used, 
#   the position at here is the node reference of the item
# 2. face 2 questions: how to transform index and element, how to build the sequence of position(index) items
# 3. when we need a component class to determine the index? because when two elements have the same value, the index cannot be determined according to the value
```
